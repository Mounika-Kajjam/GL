````python
# Load the dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
````

````python
# distplot --> distribution plots
sns.distplot(tips.total_bill)
````

````python
sns.distplot(tips.total_bill, kde =False)
````

````python
# Relplot --> relation between variables(scatter plot)
````

````python
sns.relplot('total_bill', 'tip', data = tips)
````

````python
sns.relplot('total_bill', 'tip', data = tips, hue ='day')
````

````python
sns.relplot('total_bill', 'tip', data = tips, hue ='day', row = 'smoker')
````

````python
# Categorical Plots
# catplot --> categorical plot
sns.catplot('day', 'total_bill', data=tips)
````

````python
sns.swarmplot('day', 'total_bill', data=tips)
````

````python
sns.catplot('day', 'total_bill', data=tips, kind='swarm')
````

````python
sns.catplot('day', 'total_bill', data=tips, kind='box')
#sns.boxplot
````
