```python
# sort the dataset by a column
data.sort_values(by='Age', ascending=False). head(5)
```

```python
# Access a particluar column
data.Sex
```

```python
# value_counts --> returns the frequency of each category in the column
data.Sex.value_counts()
```

```python
# isna & isnull --> check for the missing values
data.Sex.isna().sum()
```

```python
# isnull()
data.Sex.isnull().sum()
```

```python
# Displays Boolean values if any of the Age or Cabin value is missing
data.Age.isnull()| data.Cabin.isnull()
```

```python
# Display the records where the data is missing in either of the Age or Cabin columns
data[data.Age.isnull()| data.Cabin.isnull()]
```

```python
# Drops the records where the data is missing
df = data.dropna()
df.shape
```

```python
data.shape
```

```python
# fillna --> Fill the missing values with a secified value
data.fillna(0, inplace = True)
data.isnull().sum()
```

```python
# drop --> Drops the Passenger column from the dataset
# axis=1 indicates drop with respect to column
data.drop('PassengerId', inplace=True)
data.head()
```

```python
# Drop the column name ticket and store the resulting data into df1
df1 = data.drop('Ticket', axis=1)
df1.head()
```

```python
# Drop the rows with indices 2,3 and 4
data.drop([2,3,4], axis=0, inplace=True)
data.head()
```

# loc and iloc
loc--> location
iloc --> index location

```python
# Accessing the value of the row label 5 and column label Pclass
data.loc[5, 'Pclass']
```

```python
# index of which row you want to access & index of which column you want to access
# Accessing the value of the row index 5 and column index 1
data.iloc[5, 1]
```

```python
# Display a subset of the rows labels 0 to 5 and column labels Survived to Name
data.loc[0:5, "Survived":"Name"]
```

```python
# Display a subset of the rows with indices 0 to 5 and columns with indices 0 to 3
data.iloc[0:5, 0:3]
```

```python
# :--> indicates all
# Display all teh rows of the column Pclass
data.loc[:, "Pclass"]
```

###  Aggregation

```python
data.groupby('Sex')
```

```python
g = data.groupby('Sex')
for gender, gender_df in g:
    print(gender)
    print(gender_df)
```

```python
data.groupby('Sex').agg({'Age': ['min', 'max', 'count'], 'Fare': ['min', 'max']})
```
